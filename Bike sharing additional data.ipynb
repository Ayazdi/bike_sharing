Adding the weather data:

```python
import requests
from datetime import datetime
```

```python
date_range = pd.date_range('2013/1/1', '2014/12/31', freq='D').tz_localize('US/Eastern')
len(date_range)
```

```python
weather = pd.DataFrame()
for d in range(len(date_range)):
    result = requests.get(f'https://api.darksky.net/forecast/5b83fb0d8d5d0462839fba7ef5ab09ab/38.907,-77.0417,{int(date_range[d].timestamp())}?units=ca')
    results_json = result.json()
    df_1 = pd.DataFrame(columns=['time', 'temperature', 'atemperatue', 'humidity', 'windSpeed'])
    for entry in results_json['hourly']['data']:
        df_1.loc[len(df_1)] = [pd.to_datetime(entry['time'], unit='s').tz_localize('uct').tz_convert('US/Eastern'), entry['temperature'], entry['apparentTemperature'], entry['humidity'], entry['windSpeed']]
    weather = pd.concat([weather, df_1], sort=True)
```
